```python
from splinter import Browser
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
In [4]:
def init_browser():
    
    executable_path = {'executable_path': 'chromedriver.exe'}
    return Browser("chrome", **executable_path, headless=False)
In [90]:
def scrape():
    
    mars_data = {}
    browser = init_browser()
    # visit website to gather data
    site1 = "https://mars.nasa.gov/news/"
    browser.visit(site1)
    html = browser.html

    
    soup_obj = BeautifulSoup(html, "html.parser")
    news_title = soup_obj.find('li', attrs={'class': 'slide'}).find('div', attrs={'class': 'image_and_description_container'}).find('div', attrs={'class': 'list_text'}).find('div', attrs={'class': 'content_title'}).text
    article_teaser_body = soup_obj.find('li', attrs={'class': 'slide'}).find('div', attrs={'class': 'image_and_description_container'}).find('div', attrs={'class': 'list_text'}).find('div', attrs={'class': 'article_teaser_body'}).text
 
    
    mars_data["news_title"] = news_title
    mars_data["article_teaser_body"] = article_teaser_body
    
    site2 = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
    browser.visit(site2)
    html = browser.html
    
    
    soup_obj = BeautifulSoup(html, "html.parser")
    image = soup_obj.find('li', attrs={'class': 'slide'}).find('a')['data-fancybox-href']
    
    image_src = "https://www.jpl.nasa.gov"+image
    
    mars_data["featured_image_url"] = image_src
    
    site3 = "https://twitter.com/marswxreport?lang=en"
    browser.visit(site3)
    html = browser.html
   
    soup_obj = BeautifulSoup(html, "html.parser")
    tweet = soup_obj.find('div', attrs={'class': 'js-tweet-text-container'}).find('p').text
    mars_data["mars_weather"] = tweet

    browser.quit()
    
    site4 = "http://space-facts.com/mars/"
    table = pd.read_html(site4)
    df = table[0]
    df.columns = ['Description','Value']
    df.set_index('Description', inplace=True)
    mars_facts = df.to_html()
    mars_facts = mars_facts.replace('\n', '')
    mars_data["mars_facts"] = mars_data
    
    print(mars_data) 
    return mars_data
In [91]:
scrape()
{'news_title': "NASA's InSight Places First Instrument on Mars", 'article_teaser_body': 'In deploying its first instrument onto the surface of Mars, the lander completes a major mission milestone.', 'featured_image_url': 'https://www.jpl.nasa.gov/spaceimages/images/largesize/PIA22994_hires.jpg', 'mars_weather': 'Sol 2283 (2019-01-07), high -9C/15F, low -71C/-95F, pressure at 8.25 hPa, daylight 06:44-18:55', 'mars_facts': {...}}
Out[91]:
{'news_title': "NASA's InSight Places First Instrument on Mars",
 'article_teaser_body': 'In deploying its first instrument onto the surface of Mars, the lander completes a major mission milestone.',
 'featured_image_url': 'https://www.jpl.nasa.gov/spaceimages/images/largesize/PIA22994_hires.jpg',
 'mars_weather': 'Sol 2283 (2019-01-07), high -9C/15F, low -71C/-95F, pressure at 8.25 hPa, daylight 06:44-18:55',
 'mars_facts': {...}}
```
